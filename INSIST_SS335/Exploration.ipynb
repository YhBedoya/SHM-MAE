```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ss335-acc-201951.csv", index_col=0)
df
```

```python
dfs = df[df["sens_pos"]=="S6.1.2"]
dfs
```

```python
dfs['ts'] = pd.to_datetime(dfs['ts'], unit='ms')
dfs
```

```python
dfs["x"][1]
```

```python
def flatter(row):
    row["x"] = row["x"].replace("\n", "").replace("[", "").replace("]", "").split(" ")
    row["y"] = row["y"].replace("\n", "").replace("[", "").replace("]", "").split(" ")
    row["z"] = row["z"].replace("\n", "").replace("[", "").replace("]", "").split(" ")

    return row

dfs = dfs.apply(flatter, axis=1)
dfs
```
